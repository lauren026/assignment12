```python
import numpy as np
import matplotlib.pyplot as plt
```

Define the function.

```python
num     = 1001
std     = 5 

def fun(x):
 
    f = np.abs(x) * np.sin(x)

    return f
```

And set the variables to use.

```python
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x) # clean points
y2      = y1 + nn * std # noisy points
```

```python
plt.title('Clean Data')
plt.plot(x, y1, 'b.')
plt.show()
```

The above graph is about to clean data.

```python
plt.title('Noisy Data')
plt.plot(x, y2, 'r.')
plt.show()
```

And, The above graph is about to noisy data.

```python
def leastSquare(x, y, a):

    n = np.zeros((a+1, num))
    
    
    for i in range(0,a+1):
        n[i] = np.array([x ** i])
    
    n = n.T
    inv = np.dot(np.linalg.inv(np.dot(n.T, n)), n.T)
    
    return np.dot(inv, y)
```

This is the leastSquare function.

```python
for i in range(6, 16):
    k = leastSquare(x, y2, i)
    
    a = 0
    for j in range(0, i+1):
        a += k[j] * x ** j
    
    e = y1 - a
    e = e ** 2
    
    plt.title('k : %d' %i)
    plt.plot(x, a, 'k.', x, e, 'g.')
    plt.show()
```

The polynomial curves that fit the noisy data by the least square error with varying.

```python
i=[1/8, 1/4, 1/2, 1, 2, 4, 8]
for k in i:
    p = leastSquare(x, y2, k)
    
    a = 0
    for j in range(0, k+1):
        a += p[j] * x ** j
    
    err = y1 - a
    err = err ** 2
    
    plt.title('Error p : %d' %k)
    plt.plot(x, a, 'k.', x, err, 'g.')
    plt.show()
```
